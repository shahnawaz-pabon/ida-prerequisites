## Dataset Preparation
**Why this step matters:** Properly structured data avoids errors in analysis and makes regression modeling straightforward.

```r
# Simulated dataset
set.seed(123)
data <- data.frame(
  Lighting = rep(c("Warm", "Cool", "Natural"), each = 20),
  Reading_Score = c(
    rnorm(20, mean = 75, sd = 5),  # Warm lighting scores
    rnorm(20, mean = 70, sd = 5),  # Cool lighting scores
    rnorm(20, mean = 80, sd = 5)   # Natural lighting scores
  )
)

# Convert Lighting to dummy variables
data$Lighting_Dummy <- ifelse(data$Lighting == "Warm", 1, 0)
print(data)
```

## Exploratory Data Analysis (EDA)
**Why EDA matters:** EDA reveals patterns, potential issues, and the distribution of the data. This ensures model assumptions (e.g., normality) are met.

### Summary Statistics
Provide insights into central tendencies and variability in reading scores.

```r
# Install dplyr
install.packages("dplyr")

library(dplyr)

# Summary statistics
summary_stats <- data %>%
  group_by(Lighting) %>%
  summarise(
    Mean_Score = mean(Reading_Score),
    Median_Score = median(Reading_Score),
    SD_Score = sd(Reading_Score),
    Min_Score = min(Reading_Score),
    Max_Score = max(Reading_Score)
  )
print(summary_stats)
```

### Visualizations
Visualize the distribution of reading scores to identify trends or outliers.

```r
library(ggplot2)

# Boxplot for Reading Scores by Lighting Condition
ggplot(data, aes(x = Lighting, y = Reading_Score, fill = Lighting)) +
  geom_boxplot() +
  labs(
    title = "Reading Scores by Lighting Condition",
    x = "Lighting Condition",
    y = "Reading Score"
  ) +
  theme_minimal()
```

## Simple Linear Regression
**Why regression matters:** Regression quantifies the relationship between lighting conditions and reading scores, providing coefficients that estimate how scores change under Warm lighting.

### Key Insights:

- **Intercept:** Predicted reading score for Cool/Natural lighting.
- **Coefficient (Lighting_Dummy):** Change in score when switching to Warm lighting.

```r
# Simple linear regression model
model <- lm(Reading_Score ~ Lighting_Dummy, data = data)
summary(model)
```

## ANOVA Test
**Why ANOVA matters:** While regression focuses on Warm lighting, ANOVA evaluates if any lighting condition significantly influences scores.

### Insight
- A significant p-value indicates differences in mean scores between at least two lighting conditions.

```r
# One-way ANOVA
anova_model <- aov(Reading_Score ~ Lighting, data = data)
summary(anova_model)
```

## Post-Hoc Pairwise Comparison
**Why pairwise comparisons matter:** After ANOVA, this identifies specific lighting conditions with significant score differences.

```r
# Pairwise t-tests
pairwise.t.test(data$Reading_Score, data$Lighting, p.adjust.method = "bonferroni")
```

## Effect Size Analysis
**Why effect size matters:** Statistical significance doesn't always imply practical relevance. Effect size quantifies the strength of differences.

```r
install.packages("effsize")
library(effsize)

# Cohen's d for Warm vs. Natural lighting
cohen_d <- cohen.d(
  Reading_Score ~ Lighting,
  data = subset(data, Lighting %in% c("Warm", "Natural"))
)
print(cohen_d)
```

## Residual Analysis
**Why residuals matter:** Residuals check the assumptions of regression, like linearity and homoscedasticity.

```r
# Residual diagnostics
par(mfrow = c(2, 2))
plot(model)
```

## Bootstrap Analysis
**Why bootstrapping matters:** Provides robust confidence intervals for coefficients, ensuring reliability even with small samples.

```r
library(boot)

# Define bootstrapping function
boot_fn <- function(data, indices) {
  d <- data[indices, ]
  coef(lm(Reading_Score ~ Lighting_Dummy, data = d))
}

# Perform bootstrap
set.seed(123)
boot_results <- boot(data, boot_fn, R = 1000)
print(boot_results)
```

## Sensitivity Analysis
**Why sensitivity matters:** Removing outliers tests model stability.

```r
# Identify and remove outliers
outliers <- boxplot.stats(data$Reading_Score)$out
data_cleaned <- data[!data$Reading_Score %in% outliers, ]

# Re-run regression
model_cleaned <- lm(Reading_Score ~ Lighting_Dummy, data = data_cleaned)
summary(model_cleaned)
```
